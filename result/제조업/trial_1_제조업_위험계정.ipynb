## 4.1. 상하위 1%에 대해서만 없애는방법

```python
def top_removal(dataset, percentage):
    
    drop_index = set([])
    
    for col in dataset.columns[:-1]:
        
        feature = dataset[col]
        feature_name = col
        
        
        feat_upper, feat_lower = np.percentile(feature, 100-percentage), np.percentile(feature, percentage)

        outliers = [x for x in feature if x < feat_lower or x > feat_upper]
        print(feature_name + ' Amount of outliers: {}'.format(len(outliers)))
        #print(feature_name + ' outliers:{}'.format(outliers))

        
        removal_index = dataset[(dataset[feature_name] > feat_upper) | (dataset[feature_name] < feat_lower)].index
        
        drop_index.update(list(removal_index))
        
        # k.update([3, 4, 5]) : 집합에 여러 값 추가시 사용
        
        print('제거 레코드 수: %d' %len(drop_index))

        print('-' * 65)
    
    dataset = dataset.drop(list(drop_index))    
    print('남은 레코드 수: %d' %len(dataset))
    
    return dataset       
        
df_removed1 = top_removal(df,1)
```

4.3. 이상치데이터를 학습.

```python
def outliers_removal(dataset, multiplier):
    
    drop_index = set([])
    
    for col in dataset.columns[:-1]:
        
        feature = dataset[col]
        feature_name = col
        
        
        q25, q75 = np.percentile(feature, 25), np.percentile(feature, 75)
        feat_iqr = q75 - q25    
        feat_cut_off = feat_iqr * multiplier
        feat_lower, feat_upper = q25 - feat_cut_off, q75 + feat_cut_off

        outliers = [x for x in feature if x < feat_lower or x > feat_upper]
        print(feature_name + ' Amount of outliers: {}'.format(len(outliers)))
        #print(feature_name + ' outliers:{}'.format(outliers))

        
        removal_index = dataset[(dataset[feature_name] > feat_upper) | (dataset[feature_name] < feat_lower)].index
        
        drop_index.update(list(removal_index))
        
        # k.update([3, 4, 5]) : 집합에 여러 값 추가시 사용
        
        print('제거 레코드 수: %d' %len(drop_index))

        print('-' * 65)
    
    dropset = dataset.loc[drop_index]
    dataset = dataset.drop(list(drop_index))
        
    print('남은 레코드 수: %d' %len(dataset))

    return dataset, dropset
       
        
df_removed2, df_dropping2 = outliers_removal(df, 1.5)
```
